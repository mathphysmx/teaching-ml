# Linear regression

La ecuacion modelo es

$y = \theta x$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xi = -4
yi = 2

theta = yi / xi
x = np.linspace(-6, 0)
y = theta * x
y
```

```python
plt.plot(x, y)
plt.plot(xi, yi, 'r.')
```

# Linear regression ($y = \theta x$, $m = 2$)

```python
import numpy as np
import matplotlib.pyplot as plt
xi = np.array([0, 5])
yi = np.array([5, 0])
theta = sum(xi * yi) / sum(xi**2)
theta
```

```python
theta
```

```python
plt.scatter(xi,yi)
x=np.linspace(min(xi),max(xi),num=5)
y=theta*x
plt.plot(x,y)
plt.axis('equal')
```

# Linear regression ($y = \theta x$, $m$)

```python
# Fake data
m = 1000
xi = np.random.rand(m, 1)
yi = 20 * xi + np.random.randn(m, 1)
plt.scatter(xi, yi)
# model
theta = sum(xi * yi) / sum(xi**2)
theta

x = np.linspace(0, 1)
y = theta * x
plt.plot(x, y, "r-")
```
